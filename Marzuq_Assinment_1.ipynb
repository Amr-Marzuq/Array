## Remainder

```python
x, y= input('Enter 2 numbers: ').split()
x = int(x)
y = int(y)
if (x>y):
  z = x%y 
  if (z== 0):
    print('The remainder is Zero')
  else: 
    print("The remainder is ", z)
else:
  print('The first number must be larger than the second')
```

```python
x, y= input('Enter 2 numbers: ').split()
x = int(x)
y = int(y)
if (x>y):
  z = x/y
  a = int(z)
  z = z-a
  z = z*y
  z = int(z)
  if (z== 0):
    print('The remainder is Zero')
  else: 
    print("The remainder is ", z)
else:
  print('The first number must be larger than the second')
```

## Switch variables

```python
x, y= input('Enter 2 numbers: ').split()
x = int(x)
y = int(y)
z = x
x = y
y = z
print('x= ',x, 'y= ',y)
```

## Convert to seconds

```python
x,y,z = input('Enter time in H:M:S ').split(':')
x = int(x)
y = int(y)
z = int(z)
a = 3600* x + 60*y + z
print('Time in Seconds is 0:0:',a)
```
